```python
import pandas as pd
import matplotlib.pyplot as pt

%matplotlib inline  
```

```python
read_in = open("../Fitted_particles/tree_0_fit_50_particles.tsv")
```

```python
str_in = read_in.read()
```

```python
str_in
```

```python
cells = str_in.split('Tree ')[1:]
```

```python
lines
```

```python
col_map = {
    1: "Kpg",
    2: "KInhibitPu",
    3: "Kdp",
    4: "Kpp",
    5: "KInhibitGata",
    6: "Kdg"
}

data = {}
```

```python
for cell in cells:
    
    lines = cell.splitlines()

    this_tree = lines[0][0]
    this_cell = lines[0].split(" ")[2]

    this_id = this_tree + "_" + this_cell

    value_map = {
        "Kpg": [],
        "KInhibitPu": [],
        "Kdp": [],
        "Kpp": [],
        "KInhibitGata": [],
        "Kdg": []
    }

    for row in lines[2:]:
    
        cols = row.split("\t")
    
        for col_num in col_map:
        
            value_map[col_map[col_num]] += [cols[col_num]]
            
            
    this_df = pd.DataFrame(value_map)
    
    for col in this_df.columns:
        this_df[col] = this_df[col].astype("float")
        
    data[this_id] = this_df
```

```python
value_map = {
    "Kpg": [],
    "KInhibitPu": [],
    "Kdp": [],
    "Kpp": [],
    "KInhibitGata": [],
    "Kdg": []
}
```

```python
for this_id in data.keys():
    
    this_df = data[this_id]
    
    for val in value_map.keys():
        
        value_map[val] += this_df[val].tolist()
```

```python
pt.hist(value_map["Kpg"], bins=100)

f = pt.gcf()
f.set_size_inches(16,10)
pt.title("Distribution of Kpg")
pt.xlabel("Kpg")
pt.ylabel("Occurrences")

pt.show()
```
